```python
from contextily.place import *
%matplotlib inline

import contextily as cx
import geopandas as gpd
import rasterio
from rasterio.plot import show as rioshow
import matplotlib.pyplot as plt
from geodatasets import get_path
```

```python
data_url = "https://ndownloader.figshare.com/files/20232174"
db = gpd.read_file(data_url)
```

```python
ax = db.plot(color="red", figsize=(9, 9))
cx.add_basemap(ax, crs=db.crs)
plt.show()
```

```python
ax = db.plot(color="red", figsize=(9, 9))
cx.add_basemap(ax, crs=db.crs, zoom_adjust=5)
plt.show()
```
